```python
# Columns to drop
columns_to_drop = ['City', 'Address','Locality','Locality Verbose','Longitude','Latitude']

df = df.drop(columns=columns_to_drop)
df.head()
```

```python
#CHECKING IF THERE ARE ANY NULL OR MISSING VALUES IN THE DATASET
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
#LET US DO SOME STATISTICAL TECHNIQUES ON THE DATASET
```

```python
df.info()
```

```python
df.describe()
```

```python
#WE HAVE TO CHANGE THE CATEGORICAL VARIABLES INTO NUMERIC USING LABELENCODER,ONE-HOT ENCODER,ORDINALENCODER
```

```python
#LABEL ENCODING

from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()
df['Switch to order menu'] = labelencoder.fit_transform(df['Switch to order menu'])
df['Has Table booking'] = labelencoder.fit_transform(df['Has Table booking'])
df['Has Online delivery'] = labelencoder.fit_transform(df['Has Online delivery'])
df['Is delivering now'] = labelencoder.fit_transform(df['Is delivering now'])
df.head()
```

```python
df.Cuisines
```

```python
#ORDINAL ENCODING FOR RATING COLOR COLUMN

from sklearn.preprocessing import OrdinalEncoder
ordinalencoder = OrdinalEncoder(categories=[['Orange', 'White', 'Yellow','Green','Dark Green','Red']])
df['Rating color'] = ordinalencoder.fit_transform(df[['Rating color']])
df.head()
```

```python
#ORDINAL ENCODING FOR RATING TEXT COLUMN
ordinalencoder = OrdinalEncoder(categories=[['Not rated','Poor','Average','Good','Very Good','Excellent']])
df['Rating text'] = ordinalencoder.fit_transform(df[['Rating text']])
df.head()
```

```python
#I AM USING MINMAX SCALAR TO SCALE THE VALUES BETWEEN 0 AND 1 (MINMAX NORMALIZATION)

from sklearn.preprocessing import MinMaxScaler
minmax_scaler = MinMaxScaler()
df[['Country Code','Average Cost for two']] = minmax_scaler.fit_transform(df[['Country Code','Average Cost for two']])
df.head()
```

```python
#USING ONE HOT ENCODING (DUMMIES)

df_dummies = pd.get_dummies(df, columns=['Restaurant Name','Currency'], drop_first=True)
df_dummies.head()
```

```python
#CONVERTING TRUE AND FALSE TO 1 AND 0 RESPECTIVELY

df_numeric = df_dummies *1
df_numeric
```

```python
df_numeric.info()
```

```python
df_numeric['Cuisines'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df_numeric['Cuisines'] = label_encoder.fit_transform(df_numeric['Cuisines'])
df_numeric.head()
```

```python
#defining feature and target
x = df_numeric.drop('Cuisines',axis=1)
y = df_numeric['Cuisines']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, classification_report

# Calculate performance metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
class_report = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy:.2f} %")
print(f"Precision: {precision:.2f} %")
print(f"Recall: {recall:.2f} %")
print("Classification Report:\n", class_report)
```
